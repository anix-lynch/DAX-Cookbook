```python
from google.colab import files

# Upload the file
uploaded = files.upload()
```

```python
import io

# Load the CSV file into a DataFrame
salaries_df = pd.read_csv(io.BytesIO(uploaded['salaries.csv']))
```

```python
# Calculate the average salary for each job title
average_salaries = salaries_df.groupby('job_title')['salary_in_usd'].mean().reset_index()

# Sort by average salary in descending order
top_salaries = average_salaries.sort_values(by='salary_in_usd', ascending=False).head(10)

top_salaries
```
